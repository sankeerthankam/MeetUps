# Data Visualization in Python - Seaborn

## Seaborn 

A Python data visualization library built on matplotlib that helps plot attractive statistical graphs.

___
## Section I - Distribution Plots

To better understand the data, we need to understand the distribution of the data set. Seaborn provides distplot, jointplot, pairplot etc that help us understand the distributions

### Importing Seaborn and Tips dataset

```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### distplot

The distplot shows the distribution of a univariate (single variable) data.

```python
sns.distplot(tips['total_bill'])
```

You can remove the kde layer and specify the number of bins of interest

```python
sns.distplot(tips['total_bill'],kde=False,bins=25)
```

## jointplot

jointplot() plots two distplots for bivariate (two variable) data. 

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

By default, jointplot uses a scatterplot between two variable. However, you can also choose your plot of interest by using the 'kind' parameter

'kind' parameter accepts the following values: 
* scatter 
* reg 
* resid 
* kde 
* hex

```python
# Jointplot with hexagonal bins
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Jointplot with Regression and KDE
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

You can see that when you use specify 'kind' = regression, 'jointplot' automatically specifies the p-value and  pearson correlation between the two variables.

The band you see in the graph denotes the 95% confidence interval.

## pairplot

Helps plot pairwise relationships across entire dataframe (only for the numerical columns).

```python
sns.pairplot(tips)
```

However, it supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## Section II - Categorical Plots

Categorical plots helps us plot categorical data. Seaborn provides countplot, barplot, boxplot, violinplot, stripplot, swarmplot etc that help us understand categorical data better.

## countplot

Helps plot the number of occurrences of each class in a categorical variable

```python
sns.countplot(x='sex',data=tips)
```

## barplot

Similar tp countplot but helps plot aggregation of categorical data based off statistical functions such as mean, standard deviation etc.

```python
# Uses mean as an estimator by default 
sns.barplot(x='sex',y='total_bill',data=tips) 
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## boxplot

boxplots show the spread of the categorical data between variables or across classes (levels) of a categorical variable.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='Set2')
```

### Interpreting a box and whisker plot

![title](boxplot.png)

```python
# Hue variable helps differentiate different levels of the categorical variable
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

## stripplot

A strip plot is a good complement to a box plot in cases where you have small number of observations.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

It is apparent that the total_bill is below $30 but it does not show 'how' apparent it is. We can use 'jitter' = True to know the distribution 

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

## swarmplot

The swarmplot is similar to stripplot, but the points are adjusted so that they dont overlap. This gives a better representation of the distribution, although it does not scale well for large numbers of observations 

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

## violinplot

A violin plot also shows the distribution of quantitative oe categorical data across several levels. However, violinplot uses 'kernal density estimation' to show the distribution of the data whereas boxplot features the actual datapoints.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='Set1')
```

You can also use 'hue' variable as well as 'split' 

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```
